## Partie III — Inspection, Profilage et Nettoyage

### Étape 3 — Inspection des types et des valeurs manquantes

```python
commande_df.info()
print('\nValeurs manquantes dans order_details :')
print(commande_df.isna().sum())
```

```python
menu_df.info()
print('\nValeurs manquantes dans menu_items :')
print(menu_df.isna().sum())
```

### Question 4 — Type de la colonne order_date (1 pt)

```python
print(f'Type de order_date après chargement initial : {commande_df["order_date"].dtype}')
print(f'Exemple de valeur : "{commande_df["order_date"].iloc[0]}"')
```

**Réponse Q4 :**  
La colonne `order_date` est chargée avec le type **`object` (str)**, ce qui est **incorrect**. Les dates doivent être de type `datetime64[ns]` pour permettre des opérations temporelles (filtrage, resampling, extraction du jour/mois). Pandas ne convertit pas automatiquement les dates lors du chargement — il faut soit utiliser `parse_dates`, soit appeler `pd.to_datetime()` après le chargement.

### Étape 4 — Correction des types temporels

```python
# Rechargement avec parse_dates
commande_df_correct = pd.read_csv(
    '../data/raw/order_details.csv',
    parse_dates=['order_date']
)
# Conversion manuelle de l'heure (format 12h AM/PM)
commande_df_correct['order_time'] = pd.to_datetime(
    commande_df_correct['order_time'], format='mixed'
)

commande_df_correct.info()
```

```python
# Vérification des types corrigés
print('Types après correction :')
print(commande_df_correct[['order_date', 'order_time']].dtypes)
```

### Étape 5 — Traitement des valeurs manquantes
On supprime les lignes dont `item_id` est manquant, car elles ne peuvent pas être associées à un article du menu.

```python
print(f'Lignes avant nettoyage : {len(commande_df_correct)}')
print(f'Valeurs manquantes sur item_id : {commande_df_correct["item_id"].isna().sum()}')

# Suppression des lignes sans item_id
commande_df_correct.dropna(inplace=True)

# Vérification
print('\nValeurs manquantes après nettoyage :')
print(commande_df_correct.isna().sum())
print(f'\nLignes restantes : {len(commande_df_correct)}')
```

### Question 5 — Lignes après nettoyage et justification du choix (2 pts)

**Réponse Q5 :**  
Après suppression des 137 lignes avec `item_id` manquant, il reste **12 097 lignes** (12 234 − 137).  

**Justification du choix de suppression plutôt qu'imputation :**  
- La colonne `item_id` est la **clé de jointure** avec le menu. Sans elle, il est impossible de connaître l'article commandé, son prix ou sa catégorie.  
- L'imputation (ex : remplacer par la moyenne ou le mode) n'aurait pas de sens métier : un article fictif fausserait les calculs de revenus.  
- Les 137 lignes représentent environ **1,1 % du dataset**, ce qui est négligeable : leur suppression ne biaise pas significativement les analyses.  
- En revanche, si ces lignes étaient majoritaires pour une période donnée, il faudrait investiguer la cause (panne du système de caisse, erreur d'import).

## Partie IV — Jointure et Enrichissement des Données

### Étape 6 — Fusion des tables
On effectue une jointure interne (`inner join`) entre les commandes et le menu sur l'identifiant de l'article.

```python
print(f'Lignes avant merge : {len(commande_df_correct)}')

commande_df_correct = commande_df_correct.merge(
    menu_df,
    left_on='item_id',
    right_on='menu_item_id',
    how='inner'
)

print(f'Lignes après merge  : {len(commande_df_correct)}')
print(f'\nNouvelles colonnes : {commande_df_correct.columns.tolist()}')
commande_df_correct.head(3)
```

### Question 6 — Résultat de la jointure (2 pts)

**Réponse Q6 :**  
Après la jointure, le DataFrame contient **9 colonnes** au lieu de 5 :  
- Colonnes d'origine : `order_details_id`, `order_id`, `order_date`, `order_time`, `item_id`  
- **Nouvelles colonnes** issues du menu : `menu_item_id`, `item_name`, `category`, `price`  

Le nombre de lignes **reste identique (12 097)** car tous les `item_id` présents dans `order_details` après nettoyage correspondent à un article existant dans `menu_items`.  

**Fonctionnement d'un inner join :** Il ne conserve que les lignes pour lesquelles la clé de jointure existe dans **les deux tables**. Toute commande référençant un article inexistant dans le menu serait éliminée — mais ce n'est pas le cas ici.

### Étape 7 — Calcul du revenu total TTC

```python
# Taxe à 8 % arrondie à 2 décimales
commande_df_correct['tax'] = (commande_df_correct['price'] * 0.08).round(2)

# Revenu total TTC
commande_df_correct['total_revenu'] = (
    commande_df_correct['price'] + commande_df_correct['tax']
)

print('Aperçu des nouvelles colonnes :')
commande_df_correct[['item_name', 'price', 'tax', 'total_revenu']].head(5)
```

### Question 7 — Revenu total TTC (1 pt)

```python
ca_total = commande_df_correct['total_revenu'].sum().round(2)
print(f'Revenu total TTC (Jan–Mar 2023) : ${ca_total:,.2f}')
```

**Réponse Q7 :**  
Le revenu total TTC généré sur les 3 mois est de **$171 981,78**.

### Étape 10 — Revenu par catégorie

## Partie VI — Visualisation et Recommandations

### Exercice 1 — Évolution mensuelle du chiffre d'affaires (3 pts)
On utilise `resample('ME')` pour agréger le CA par mois sur l'index temporel.

```python
import matplotlib.ticker as mticker

monthly_ca = (
    commande_df_correct
    .set_index('order_date')
    .resample('ME')['total_revenu']
    .sum()
)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(monthly_ca.index, monthly_ca.values, marker='o', linewidth=2.5,
        color='steelblue', markersize=8)
ax.fill_between(monthly_ca.index, monthly_ca.values, alpha=0.1, color='steelblue')

# Annotations des valeurs
for x, y in zip(monthly_ca.index, monthly_ca.values):
    ax.annotate(f'${y:,.0f}', xy=(x, y), xytext=(0, 12),
                textcoords='offset points', ha='center', fontsize=10, fontweight='bold')

ax.set_title('Évolution du CA mensuel — Bella Vista (Jan–Mar 2023)', fontsize=14, fontweight='bold')
ax.set_xlabel('Mois', fontsize=12)
ax.set_ylabel('Revenu TTC ($)', fontsize=12)
ax.set_ylim(0, monthly_ca.max() * 1.2)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%B %Y'))
ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../visuals/ca_mensuel.png', dpi=150)
plt.show()
print('Graphique sauvegardé : visuals/ca_mensuel.png')
```

**Lecture du graphique :**  
- Janvier : $58 131 | Février : $54 862 | Mars : $58 988  
- Légère baisse en février (mois plus court, 28 jours), puis rebond en mars. Le CA est relativement stable sur la période.

### Exercice 2 — Classement des articles par revenu (4 pts)

### Exercice 3 — Heatmap Jour × Heure (4 pts)
Table pivot croisant le jour de la semaine et l'heure, avec le CA moyen comme valeur.

```python
# Extraction des colonnes temporelles
commande_df_correct['hour'] = commande_df_correct['order_time'].dt.hour
commande_df_correct['day_name'] = commande_df_correct['order_date'].dt.day_name()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

pivotable = commande_df_correct.pivot_table(
    values='total_revenu',
    index='day_name',
    columns='hour',
    aggfunc='mean'
).reindex(day_order)

plt.figure(figsize=(16, 6))
sns.heatmap(
    pivotable.round(),
    annot=True, fmt='g', cmap='vlag',
    linewidths=0.5, linecolor='white',
    cbar_kws={'label': 'CA moyen ($)'}
)
plt.title('CA moyen par Jour de la semaine et Heure de commande', fontsize=13, fontweight='bold')
plt.xlabel('Heure de la journée', fontsize=11)
plt.ylabel('Jour de la semaine', fontsize=11)
plt.tight_layout()
plt.savefig('../visuals/heatmap_jour_heure.png', dpi=150)
plt.show()
print('Graphique sauvegardé : visuals/heatmap_jour_heure.png')
```

### Question 11 — Analyse de la heatmap (3 pts)

```python
# Analyse quantitative de la heatmap
print('Heure la plus productive (moyenne sur tous jours) :')
print(pivotable.mean().nlargest(3).round(2))

print('\nJour le plus productif (moyenne sur toutes heures) :')
print(pivotable.mean(axis=1).sort_values(ascending=False).round(2))

print('\nPlages à très faible activité (heures avec peu/pas de données) :')
print(pivotable.isna().sum())
```

**Réponse Q11 :**  

**Plages horaires les plus rentables :**  
Les heures de **11h–14h (déjeuner)** et **17h–22h (dîner)** génèrent les CA moyens les plus élevés. Le pic de soirée autour de **19h–22h** est particulièrement fort, notamment le weekend.

**Jours les plus chargés :**  
Le **samedi** présente le CA moyen le plus élevé sur l'ensemble des tranches horaires, suivi du vendredi et dimanche. Les jours de milieu de semaine (mardi, mercredi) sont plus calmes.

**Créneaux à très faible activité :**  
Les plages **avant 11h** et **après 22h** montrent peu ou pas d'activité en semaine. Ces créneaux pourraient justifier une **fermeture anticipée** les soirs de semaine ou une ouverture plus tardive, permettant des économies sur les charges de personnel sans impact significatif sur le CA.

### Question 12 — 3 Recommandations opérationnelles (4 pts)

**Réponse Q12 — Recommandations pour le gérant de Bella Vista**

---

#### 🍝 Recommandation 1 — Capitaliser sur la cuisine italienne
**Observation :** La catégorie Italian est la plus rentable ($53 427, panier moyen $18,12). Le Korean Beef Bowl est l'article n°1 en revenus ($11 401), et les plats italiens (Spaghetti & Meatballs, Fettuccine Alfredo) occupent le podium.  
**Action :** Mettre en avant ces plats en tête de carte et dans les communications (réseaux sociaux, ardoise du jour). Envisager l'élargissement de l'offre italienne avec 1 ou 2 nouveaux plats premium (risotto, osso buco) pour attirer des clients à fort panier. Pour le Korean Beef Bowl, créer un "formule signature" incluant une entrée et une boisson pour augmenter le ticket moyen.

---

#### 🗓️ Recommandation 2 — Optimiser les horaires d'ouverture et la gestion du personnel
**Observation :** La heatmap révèle que le CA est concentré sur deux créneaux : 11h–14h (déjeuner) et 17h–22h (dîner). Avant 11h et après 22h, l'activité est quasi nulle en semaine. Le samedi est nettement le jour le plus rentable.  
**Action :** Réduire le personnel les matins de semaine (arrivée à 10h30 au lieu de 9h) et envisager une fermeture à 21h30 les lundis et mardis (activité faible en soirée ces jours-là). En contrepartie, renforcer les équipes le samedi soir où la demande est maximale. Cette réorganisation permettrait de réduire les charges salariales tout en maintenant la qualité de service aux heures de pointe.

---

#### 🗑️ Recommandation 3 — Réviser ou retirer les articles sous-performants
**Observation :** Chicken Tacos ($1 588), Potstickers ($1 993) et Chips & Guacamole ($2 304) représentent moins de 5% du CA total à eux trois, sur 3 mois. Ces articles mobilisent du stock, de la préparation et de l'espace en carte.  
**Action :** Mener une analyse complémentaire sur le **taux de commande** de ces articles (volume vs revenu). Si le volume est faible, les retirer du menu régulier et les proposer uniquement comme "spécial du jour" pour tester leur attractivité. Si le volume est fort mais le revenu faible (prix trop bas), envisager une légère revalorisation tarifaire. Simplifier le menu permet également de réduire les coûts d'approvisionnement et d'améliorer la consistance des préparations.
